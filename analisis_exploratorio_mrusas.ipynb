#### 5 - Hacer una pregunta sobre el dataset
##### *¿Cuales son las ubicaciones donde se encuentran las montañas rusas mas rapidas? (Minimo en 10 ubicaciones)*


```python
num_coasters = df['Location'].value_counts() # Cuantas veces aparece las montañas rusas en cada ubicacion
num_coasters.head()
```

```python
# FILTRAR LAS FILAS DONDE SE CUMPLA ESA CONDICION EN LA NUEVA COLUMNA
num_coasters = num_coasters[num_coasters >= 10] 
num_coasters
        
```

```python
# Obtener un valor exacto de la columna modificada 
num_coasters[num_coasters.index == 'Kings Dominion']
```

```python
"""
-Filtrar la columna 'Location' para quitar "Other"
-Agrupar el promedio (mean) y la cantidad de registros (count) de la velocidad (Speed_mph) por cada ubicación 
-Mantener solo ubicaciones con 10 o mas M.Rusas
-Ordenar de menor a mayor velocidad promedio (sort_values)
"""

ax = df.query('Location != "Other"').groupby('Location')['Speed_mph'] \
            .agg(['mean', 'count']) \
            .query('count >= 10') \
            .sort_values('mean')['mean'] \
            .plot(kind='barh', figsize=(12, 5), title='Promedio de Velocidad de las M.Rusas por Ubicacion')
ax.set_xlabel("Velocidad Promedio")
ax.set_ylabel("Ubicaciones")
plt.show()
```
